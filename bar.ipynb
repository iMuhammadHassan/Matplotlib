```python
import matplotlib.pyplot as plt

days = ['Mon','Tue','Wed','Thur','Fri','Sat','Sun']
water_intake =[2.3,3,1,2.3,4,2.3,3.9]

plt.title("daily Water intake Over a  Week")
plt.xlabel('Day')
plt.ylabel('Water Intake (Liters)')
plt.grid(True)


plt.plot(days,water_intake,linestyle='--',marker='o',color='red')
plt.savefig("waterIntake.png",dpi=300,bbox_inches='tight')
```
