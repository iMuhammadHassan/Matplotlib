```python
import matplotlib.pyplot as plt

categories = ["Rent", "Groceries", "Utilities", "Transportation", "Entertainment", "Healthcare", "Savings", "Miscellaneous"]
expenses = [1200, 350, 150, 100, 200, 180, 400, 120]

colors = ['#fb7b77', '#fdc170', '#f3f87f', '#98f786', 
          '#69ebfc', '#6d9efc', '#937df8', '#f78ef0']


plt.pie(expenses,labels=categories,autopct='%1.1f%%',colors =colors)
plt.title('Spending Breakdown by Category')
plt.savefig('pie_Chart.png',dpi=300,bbox_inches='tight')
plt.show()
```
